```python
# Importing the libaries
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Analyst – Ridgeline Plot
**Description**: A ridgeline plot shows the distribution of a continuous variable across multiple categories with overlapping density plots.

**Relevant Data**: A data analyst could use this to display the distribution of sales or customer satisfaction scores across different geographic regions or customer segments.




Loading the dataset

```python
# Load the dataset
person_data = pd.read_csv('cces_example_data.csv')
```

```python
person_data.head()
```


Creating empty figures that contains two axes stored as a matrix using the methods and syntax demonstrated in class\
Setting the title for the first axes as "Age Histogram"\
Setting the title for the second axes as "Political Identity"

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Set titles for the axes
axs[0].set_title('Age Histogram')
axs[1].set_title('Political Identity')

# Show the empty plot
plt.tight_layout()
plt.show()
```

Creating a figure that includes one axes that displays a bar plot of the frequencies of responses in the Evaluation variable\

```python
# Summarize and plot the frequencies of the Evaluation variable
evaluation_counts = person_data['Evaluation'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
evaluation_counts.plot(kind='bar', ax=ax)
ax.set_title('Frequencies of Evaluation Responses')
ax.set_xlabel('Evaluation')
ax.set_ylabel('Frequency')

plt.show()
```

Creating a figure that includes one axes that displays a scatter plot of the `ConMediaUse` and `ImmiPhobia` variables

```python
# Scatter plot for ConMediaUse and ImmiPhobia
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(person_data['ConMediaUse'], person_data['ImmiPhobia'])
ax.set_title('ConMediaUse vs. ImmiPhobia')
ax.set_xlabel('ConMediaUse')
ax.set_ylabel('ImmiPhobia')

plt.show()
```

##### Improvements for Scatterplot

**Facet or Group by Categories**: Create small multiples of scatterplots, with each one representing a different category or group (e.g., gender, income bracket). This method helps break down the data for more detailed insights within subgroups.\
**Change Marker Styles**: Adjusting the marker style or shape can help distinguish clusters of points.
